# 1. Business Understanding

The purpose of this notebook is to load the cleaned datasets into MySQL to build a relational data warehouse.

The data warehouse will support SQL analysis, business reporting, and Power BI dashboards.

The tables will be loaded while maintaining relationships between the fact table and dimension tables.

# 2. Import Libraries and Load Dataset

```python
import pandas as pd
from sqlalchemy import create_engine
customers = pd.read_csv("../data/cleaned/customers_clean.csv")
products = pd.read_csv("../data/cleaned/products_clean.csv")
categories = pd.read_csv("../data/cleaned/categories_clean.csv")
subcategories = pd.read_csv("../data/cleaned/subcategories_clean.csv")
territories = pd.read_csv("../data/cleaned/territories_clean.csv")
calendar = pd.read_csv("../data/cleaned/calendar_clean.csv")
fact_sales = pd.read_csv("../data/cleaned/fact_sales.csv")
returns = pd.read_csv("../data/cleaned/returns_clean.csv")
```

# 3. Connect MySQL

```python
engine = create_engine(
    "mysql+pymysql://root:9599846057@127.0.0.1:3306/adventureworks"
    )
```

```python
customers.to_sql(
    "customers",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)

categories.to_sql(
    "categories",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)

territories.to_sql(
    "territories",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)

calendar.to_sql(
    "calendar",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)

subcategories.to_sql(
    "subcategories",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)

products.to_sql(
    "products",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)

fact_sales.to_sql(
    "fact_sales",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)

returns.to_sql(
    "returns",
    engine,
    if_exists="append",
    index=False,
    method="multi"
)
```
