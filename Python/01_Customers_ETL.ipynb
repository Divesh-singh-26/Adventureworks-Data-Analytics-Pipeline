# 4. Data Cleaning

````python
customers.head()
customers.shape
````

````python
customers.columns = (
    customers.columns
    .str.strip()
    .str.replace(" ","_")
    .str.replace(r"(?<!^)(?=[A-Z])", "_", regex=True)
    .str.lower()
    )
````

````python
customers.columns
````

````python
customers['birth_date'] = pd.to_datetime(
    customers['birth_date'],
    errors='coerce'
    )
````

````python
customers.info()
````

````python
customers.head()
````

````python
customers['annual_income'] = (
    customers['annual_income']
    .str.replace(r'[$,]',"",regex=True)
)
````

````python
customers['annual_income'] = pd.to_numeric(
    customers['annual_income'],
    errors='coerce'
    )
````

````python
customers.info()
````

````python
customers['prefix'] = (
    customers['prefix']
    .fillna('Unknown')
)
````

````python
customers['gender'] = (
    customers['gender']
    .fillna('Unknown')
)
````

````python
customers[["prefix", "gender"]].isnull().sum()
````

````python
customers.dtypes
````

````python
text_columns = customers.select_dtypes(include='object').columns

for col in text_columns:
    customers[col]=customers[col].str.strip()
````

````python
customers['first_name'] = customers['first_name'].str.title().astype('string')
customers['last_name'] = customers['last_name'].str.title().astype('string')
````

````python
customers['marital_status'] = customers['marital_status'].replace({"M":"Married","S":"Single"}).astype("category")
````

````python
customers['gender']= customers['gender'].replace({'M':'Male','F':'Female'}).astype("category")
````
