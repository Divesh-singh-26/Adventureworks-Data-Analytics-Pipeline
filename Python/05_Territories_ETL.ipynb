# 4. Data Cleaning

```python
territories.head()
```

```python
territories.columns = territories.columns.str.strip().str.lower()
```

```python
territories = territories.rename(columns={'salesterritorykey':'sales_territory_key'})
```

```python
text_columns = territories.select_dtypes('object').columns

for col in text_columns:
    territories[col] = territories[col].str.strip() 
```
