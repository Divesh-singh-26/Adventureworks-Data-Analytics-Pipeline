# 4. Data Cleaning

```python
returns.columns = (
    returns.columns
    .str.strip()
    .str.replace(" ","_")
    .str.replace(r'(?<!^)(?=[A-Z])',"_",regex=True)
    .str.lower()
)
```

```python
returns.columns
```

```python
returns['return_date'] = pd.to_datetime(returns['return_date'],errors='coerce')
```
