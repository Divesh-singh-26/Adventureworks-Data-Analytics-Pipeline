# 5. Feature Engineering

Additional date attributes are created from the `date` column to support time-based analysis in SQL and Power BI.

These derived columns improve reporting by enabling analysis by year, quarter, month, week, day, and weekend.

```python
calendar['year'] = calendar['date'].dt.year
calendar['quarter'] ='Q' + calendar['date'].dt.quarter.astype(str)
calendar['month'] = calendar['date'].dt.month_name()
calendar['month_num'] = calendar['date'].dt.month
calendar['day'] = calendar['date'].dt.day
calendar['day_name']= calendar['date'].dt.day_name()
calendar['week'] = calendar['date'].dt.isocalendar().week.astype(int)
calendar['day_of_year']= calendar['date'].dt.day_of_year
calendar['is_weekend'] = calendar['day_name'].isin(["Saturday","Sunday"])
```
