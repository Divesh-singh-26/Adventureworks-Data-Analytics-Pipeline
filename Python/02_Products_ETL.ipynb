# 4. Data Cleaning

```python
products.columns = (
    products.columns
    .str.strip()
    .str.replace(" ","_")
    .str.replace(r'(?<!^)(?=[A-Z])',"_",regex=True)
    .str.lower()
)
```

```python
products.columns
```

```python
products = products.rename(columns={'product_s_k_u':'product_sku'})
```

```python
products['product_color']= products['product_color'].fillna('Unknown')
```

```python
products.isnull().sum()
```

```python
text_columns = products.select_dtypes(include='object').columns

for col in text_columns:
    products[col]=products[col].str.strip()
```
